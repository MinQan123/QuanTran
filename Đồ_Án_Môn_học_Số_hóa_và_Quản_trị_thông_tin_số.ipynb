```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Tạo dữ liệu giả lập số ca tai nạn giao thông TP.HCM 2014-2024
data = {
    'Năm': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Số ca tai nạn': [25.000, 22.409, 21.589, 20.080, 18.232, 18.232, 14.510, 11.495, 23.352, 22.067, 23.484]
}
df = pd.DataFrame(data)



# Phương pháp thay thế: điền bằng trung bình động 3 năm gần nhất
df['Số ca tai nạn_ma'] = df['Số ca tai nạn'].rolling(3, min_periods=1).mean().shift(1).fillna(method='bfill')

# Vẽ biểu đồ sau khi điền giá trị
plt.subplot(1, 1, 1)
plt.bar(df['Năm'], df['Số ca tai nạn'], color='#91A8ED', alpha=0.7)
plt.title('Số ca tai nạn giao thông TP.HCM từ năm 2014 đến hết năm 2024')
plt.xlabel('Năm')
plt.xticks(df['Năm'], rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("\nGiải thích: Phương pháp nội suy tuyến tính được chọn vì phù hợp với dữ liệu theo thời gian, "
      "phản ánh xu hướng thay đổi giữa các năm. Trung bình động cũng là phương án tốt "
      "nhưng có thể làm mất đi biến động tự nhiên của dữ liệu.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu số ca tai nạn giao thông TP.HCM 2014-2024
data = {
    'Năm': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Số ca tai nạn': [25000, 22409, 21589, 20080, 18232, 18232, 14510, 11495, 23352, 22067, 23484]
}
df = pd.DataFrame(data)

def display_accident_data():
    print("\nCHƯƠNG TRÌNH TRA CỨU SỐ CA TAI NẠN GIAO THÔNG TP.HCM (2014-2024)")
    print("-----------------------------------------------------------")

    while True:
        try:
            # Nhập năm từ người dùng
            year = int(input("Nhập năm (2014-2024) để tra cứu (hoặc nhập 0 để thoát): "))

            if year == 0:
                print("Cảm ơn bạn đã sử dụng chương trình!")
                break

            if year < 2014 or year > 2024:
                print("Vui lòng nhập năm trong khoảng 2014-2024!")
                continue

            # Lấy dữ liệu năm được chọn
            selected_data = df[df['Năm'] == year]
            accident_count = selected_data['Số ca tai nạn'].values[0]

            # Hiển thị kết quả
            print(f"\nNăm {year}: Số ca tai nạn giao thông là {accident_count:,} ca")

            # Vẽ biểu đồ
            plt.figure(figsize=(10, 5))

            # Biểu đồ cả chuỗi thời gian
            plt.subplot(1, 2, 1)
            plt.plot(df['Năm'], df['Số ca tai nạn'], marker='o', linestyle='-', color='blue')
            plt.title('Số ca tai nạn giao thông TP.HCM\n(2014-2024)')
            plt.xlabel('Năm')
            plt.ylabel('Số ca tai nạn')
            plt.grid(True)

            # Biểu đồ năm được chọn
            plt.subplot(1, 2, 2)
            colors = ['lightgray'] * len(df)
            colors[year - 2014] = 'red'  # Highlight năm được chọn
            plt.bar(df['Năm'], df['Số ca tai nạn'], color=colors)
            plt.title(f'Năm {year} được chọn')
            plt.xlabel('Năm')
            plt.xticks(rotation=45)

            plt.tight_layout()
            plt.show()

        except ValueError:
            print("Vui lòng nhập số năm hợp lệ!")

# Chạy chương trình
display_accident_data()
```
